```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df[['Order Date','Ship Date']].head()
```

```python
type(df["Order Date"][0])
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df["Month"]=df["Order Date"].dt.month
df["MonthName"]=df["Order Date"].dt.month_name()
df["Day"]=df["Order Date"].dt.day
```

```python
df[['Order Date','Year','Month','MonthName',"Day"]].head()
```

```python
df.duplicated().sum()
```

```python
total_sales=df["Sales"].sum()
total_profit=df["Profit"].sum()
total_sales,total_profit
```

```python
df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure()

sns.barplot(
    x = df.groupby('Category')['Profit'].sum().index,
    y = df.groupby('Category')['Profit'].sum().values
)

plt.title("Profit by Category")
plt.show()
```

```python
total_sales, total_profit
```

```python
monthly_sales=df.groupby(["Year","Month"])["Sales"].sum().reset_index()
monthly_sales.head()

```

```python
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
city_sales
```

```python
plt.figure()
city_sales.plot(kind='bar')
plt.title("Top 10 Cities by Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_product=df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
top_product
```

```python
plt.figure()
top_product.plot(kind="bar")
plt.title("Top 10 Products by Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
loss_products = df.groupby('Product Name')['Profit'].sum().sort_values().head(10)
loss_products
```

```python
plt.figure()
loss_products.plot(kind='bar')
plt.title("Top 10 Loss Making Products")
plt.ylabel("Total Profit (Loss)")
plt.xticks(rotation=90)
plt.show()
```

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
plt.figure()
top_customers.plot(kind='bar')
plt.title("Top 10 Customers by Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```
